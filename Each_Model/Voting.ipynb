```python
import pandas as pd
import numpy as np

bert = pd.read_csv("./output/Bert_submission_cv0694797761.csv",header=None)
roberta = pd.read_csv("./output/Roberta_submission_cv0704686192.csv",header=None)
deberta = pd.read_csv("./output/Deberta_submission_cv0723507720.csv",header=None)

column = pd.read_csv("./output/Exp001-roberta-base-epoch10/preds/submission_pp.csv",header=None)


bert["bert"] = bert[1]
bert["roberta"] = roberta[1]
bert["deberta"] = deberta[1]
bert["roberta_2"] = column[1]

ens = bert[["bert","roberta","deberta","roberta_2"]]

preds_data = pd.DataFrame(np.zeros((len(bert), 4)))
each_preds = pd.DataFrame(ens)

preds_data.iloc[:, 0] = (each_preds == 1).sum(axis=1)
preds_data.iloc[:, 1] = (each_preds == 2).sum(axis=1)
preds_data.iloc[:, 2] = (each_preds == 3).sum(axis=1)
preds_data.iloc[:, 3] = (each_preds == 4).sum(axis=1)

pred = np.argmax(np.array(preds_data), axis=1)
bert["ens"] = pred + 1

bert[[0,"ens"]].to_csv("./output/Emsamble.csv",index=None,header=None)
bert[[0,"ens"]]
```

```python
import pandas as pd
import numpy as np

bert = pd.read_csv("./output/Bert_submission_cv0694797761.csv",header=None)
roberta = pd.read_csv("./output/Roberta_submission_cv0704686192.csv",header=None)
deberta = pd.read_csv("./output/Deberta_submission_cv0723507720.csv",header=None)

column = pd.read_csv("./output/Exp001-roberta-base-epoch10/preds/submission_pp.csv",header=None)


bert["bert"] = bert[1]
bert["roberta"] = roberta[1]
bert["deberta"] = deberta[1]
bert["roberta_2"] = column[1]

ens = bert[["bert","roberta","deberta","roberta_2"]].T

mode = ens.mode().T

bert["voting"] = mode[0].astype(int)

bert[[0,"voting"]].to_csv("./output/Emsamble.csv",index=None,header=None)
bert[[0,"voting"]]
```
